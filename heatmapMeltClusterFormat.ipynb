```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

#Load Data

```python
matrix_df = pd.read_csv('/Users/nathanielblack/Downloads/MatrixGenomeDt_Mt_T5.txt', sep ='\t', index_col=False)
matrix_df.head()
```

#Create Sub-matrices, cluster, and then recombine

```python
def create_sub_matrix(main_df, pop, k = 10):
    sub_df = main_df[main_df['population'] == pop].copy(deep=True)
    selected_cols = [col for col in main_df.columns if col not in ['Mut','population']]
    sub_df_arr = sub_df.ix[:,selected_cols].values
    km = KMeans(n_clusters=k)
    km.fit(sub_df_arr)
    labels = km.labels_
    sub_df['group'] = labels
    sub_df.sort('group', inplace=True)
    sub_df.drop('group', axis=1, inplace=True)
    return sub_df

populations = matrix_df['population'].unique()
populations = [pop for pop in populations if pop != 'ACB']
output_matrix = create_sub_matrix(matrix_df, 'ACB')

for item in populations:
    output_matrix = output_matrix.append(create_sub_matrix(matrix_df, item), ignore_index=True)
```
